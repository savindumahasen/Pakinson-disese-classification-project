```python
from imblearn.over_sampling import SMOTE

smote =SMOTE()
x_smote,y_smote=smote.fit_resample(x,y)
print(x_smote.shape,y_smote.shape)
```

```python
y_smote.value_counts()
```

```python
## split the dataset into trainset and test set
```

```python
X_train,X_test,y_train,y_test=train_test_split(x_smote,y_smote,test_size=0.1, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
## Train the model and evaluate the performance of them
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf=RandomForestClassifier()
rf_clf.fit(X_train,y_train)
```

```python
## Make the predictions
```

```python
predictions=rf_clf.predict(X_test)
predictions
```

```python
y_test
```

```python
## evaluate the model peromance
```

```python
## Rf_clf model accuracy
```

```python
print("Rf_clf accuracy: ", accuracy_score(y_test,predictions))
```

```python
## Rf_clf model classification report
```

```python
print("Rf_clf classification_report: ", classification_report(y_test,predictions))
```

```python
## Rf_clf model confusion matrix
```

```python
print(confusion_matrix(y_test,predictions))
```
